# Harmonic oscillator with unharmonicity

A particle with mass $m$ is subject to the Harmonic potential in the x direction with constant $k$. In this case N=3.

$$K=\frac{p_x^2}{2m}+\frac{p_y^2}{2m}+\frac{p_z^2}{2m}=\frac{p^2}{2m}$$

$$U=\frac{1}{2}kq_x^2+\varepsilon q_x^3$$

Therefore the Hamiltonian ($H=K+U$) is:

$$H(q_x,q_y,q_z,p_x,p_y,p_z)=\frac{p_x^2}{2m}+\frac{p_y^2}{2m}+\frac{p_z^2}{2m}+\frac{1}{2}kq_x^2+\varepsilon q_x^3$$

The equations of motion are:

$$\left\{
\begin{array}{l}
\frac{d q_x}{d t}= \frac{p_x}{m} \\
\frac{d q_y}{d t}= \frac{p_y}{m} \\
\frac{d q_z}{d t}= \frac{p_z}{m} \\
\frac{d p_x}{d t}= -kq_x - 3 \varepsilon q_x^2\\
\frac{d p_y}{d t}= 0 \\
\frac{d p_z}{d t}= 0 
\end{array}\right.$$

Let's solve the problem considering that $m=1$ Kg, $k=1$ $N/m$, and the following initial conditions

$$\mathbf{q}_0=(0,0,0)$$

$$\mathbf{p}_0=(10,0,0)$$

Consider $\varepsilon = 0.005, 0.01, 0.02$.

```python
import numpy as np
from scipy import integrate

m = 1
k = 1
epsilon = 0.005
var0 = (0,0,0,10,0,0)

def deriv(t,var,m,k):
    qx, qy, qz, px, py, pz = var 
    return (px/m, py/m, pz/m, -k*qx -3*epsilon*qx*qx, 0, 0)

t_span = (0,10)
t = np.linspace(0,10,1000)

sol = integrate.solve_ivp(deriv, t_span, var0, t_eval=t, args=(m,k))
```

```python
import matplotlib.pyplot as plt

plt.plot(sol.t,sol.y[0],label=r"q_x")
plt.plot(sol.t,sol.y[1],label=r"q_y")
plt.plot(sol.t,sol.y[2],label=r"q_z")
plt.xlabel("t [s]")
plt.ylabel("q [m]")
plt.legend()
```

```python
import matplotlib.pyplot as plt

plt.plot(sol.t,sol.y[3],label=r"p_x")
plt.plot(sol.t,sol.y[4],label=r"p_y")
plt.plot(sol.t,sol.y[5],label=r"p_z")
plt.xlabel("t [s]")
plt.ylabel("p [Kg m/s]")
plt.legend()
```

```python
import matplotlib.pyplot as plt

omega = np.sqrt(k/m)

plt.plot(sol.t,sol.y[0],label=r"q_x",marker=".")
plt.plot(sol.t,sol.y[3,0]*np.sin(omega*sol.t),label=r"analytical harmonic")
plt.xlabel("t [s]")
plt.ylabel("q [m]")
plt.legend()
```
